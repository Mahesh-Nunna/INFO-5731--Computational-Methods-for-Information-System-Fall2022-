(2) Write a Python program to count the number of characters in a string. (4 points)

Sample String : 'google.com‘, 

Expected Result : {'o': 3, 'g': 2, '.': 1, 'e': 1, 'l': 1, 'm': 1, 'c': 1}

```python
# write your answer here
def count_char(str1):
    dict = {}
    for n in str1:
        keys = dict.keys()
        if n in keys:
            dict[n] += 1
        else:
            dict[n] = 1
    return dict
course_name= "computational methods for information systems"
print('String Given:',course_name)
print('No of Characters:',count_char(course_name))




```

(4) Write a Python program to get a string from a given string where all occurrences of its first char have been changed to '\$', except the first char itself. (4 points)

Sample String : 'restart‘, 

Expected Result : 'resta\$t'

```python
# write your answer here
def change_char(str1):
  first_char = str1[0]
  str1 = str1.replace(str1[0], '$')
  str1 = first_char + str1[1:]

  return str1


print(change_char('glowing'))
```

(10) Write a Python program to find the duplicate elements in a given array of integers. Return -1 If there are no such elements. (4 points)

```python
# write your answer here
def duplicate_ele(nums):
    new_set = set()
    no_duplicate = -1

    for i in range(len(nums)):

        if nums[i] in new_set:
            return nums[i]
        else:
            new_set.add(nums[i])

    return no_duplicate
ele_1 = [1,2.2,3,3,4.2,5,7]
ele_2 = [0.7,7.5,8.9,8.9,10]
ele_3 = [5,72,98,2,1,1,9,9,9,9]
ele_4 = [6,7,8,9,10]
print("Elements in list 1:", ele_1,"\nDuplicate Element :", duplicate_ele(ele_1))
print("Elements in list 2:", ele_2,"\nDuplicate Element :", duplicate_ele(ele_2))
print("Elements in list 3:", ele_3,"\nDuplicate Element :", duplicate_ele(ele_3))
print("Elements in list 4:", ele_4,"\nDuplicate Element :", duplicate_ele(ele_4))




```

(11) Write a Python program to select all the Sundays of a specified year. (4 extra points)

```python
# write your answer here
from datetime import date, timedelta

def sundays_in_year(year):
       #year,month,date
       dt = date(year, 1, 1)
       
       dt += timedelta(days = 6 - dt.weekday())  
       
       while dt.year == year:
          yield dt
          dt += timedelta(days = 7)
no_of_weeks=0  
for sundays in sundays_in_year(1998):
   no_of_weeks +=1
   print(sundays,"Sunday")
   
print("Total No Of Sundays:",no_of_weeks)





```
