Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python
# You code here (Please add comments in the code):

import requests
from bs4 import BeautifulSoup
import pandas as pd
Rating = [] # List to store Review headings
Review =[] #List to store reviews
Rev_place =[]

#Collecting the Harry Potter Box Set: The Complete Collection (Children's Paperback) Reviews from Flipkart
for number in range(100):
  requested_link = "https://www.flipkart.com/harry-potter-box-set-complete-collection-children-s-paperback/product-reviews/itma23355d0b5e66?pid=9781408856772&lid=LSTBOK9781408856772SQ5QLG&marketplace=FLIPKART" + str(number) # Generating link dynamically
  #Getting Request for the link above
  web_page = requests.get(requested_link)
  soup = BeautifulSoup(web_page.text, 'html.parser')
  rating_reviews = soup.find_all(class_='_3LWZlK _1BLPMq')
  text_reviews = soup.find_all(class_='_2-N8zT')
  review_place =soup.find_all(class_='_2mcZGG')
  for ele, sub_ele,rev_pl in zip(rating_reviews, text_reviews, review_place) : # Iterating through the list
      Rating.append(ele.text) #Appending to empty list
      Review.append(sub_ele.text)
      Rev_place.append(rev_pl.text)

#Dataframe Creation by combining collected data and making the column names of our own!
harry_potter = pd.DataFrame(list(zip(Rating, Review,Rev_place )), columns =['Rating', 'Short Review', 'Place']) 
print("Total Data Frame Length is {0}".format(len(harry_potter)))

harry_potter
```
