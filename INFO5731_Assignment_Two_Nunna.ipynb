(40 points). Write a python program to collect text data from **either of the following sources** and save the data into a **csv file**:

(1) Collect all the customer reviews of the product [Apple iPhone 11](https://www.amazon.com/Apple-iPhone-11-64GB-Unlocked/dp/B07ZPKF8RG/ref=sr_1_13?dchild=1&keywords=iphone+12&qid=1631721363&sr=8-13) on amazon.

(2) Collect the top 10000 User Reviews of the film [Shang-Chi and the Legend of the Ten Rings](https://www.imdb.com/title/tt9376612/reviews?ref_=tt_sa_3) from IMDB.

(3) Collect all the reviews of the top 100 most popular software from [G2](https://www.g2.com/) or [Capterra](https://www.capterra.com/)

(4) Collect the abstracts of the top 10000 research papers by using the query [natural language processing](https://citeseerx.ist.psu.edu/search?q=natural+language+processing&submit.x=0&submit.y=0&sort=rlv&t=doc) from CiteSeerX.

(5) Collect all the information of the 904 narrators in the [Densho Digital Repository](https://ddr.densho.org/narrators/).

(6) Collect the top 10000 tweets by using hashtag ["#blacklivesmatter"](https://twitter.com/hashtag/blacklivesmatter) from Twitter. 


```python
# You code here (Please add comments in the code):

import requests
from bs4 import BeautifulSoup
import pandas as pd
Rating = [] # List to store Review headings
Review =[] #List to store reviews
Comp_rev =[]
Rev_place =[]


#collecting the IMDB review from the IMDb website for the film Shang-Chi and the Legend of the Ten Rings.
for number in range(400):
  requested_link = "https://www.imdb.com/title/tt9376612/reviews?ref_=tt_sa_3" + str(number) # Generating link dynamically
  #Getting Request for the link above
  web_page = requests.get(requested_link)
  soup = BeautifulSoup(web_page.text, 'html.parser')
  rating_reviews = soup.find_all(class_='rating-other-user-rating')
  text_reviews = soup.find_all(class_='title')
  total_review = soup.find_all(class_='text') 
  review_date =soup.find_all(class_='review-date')

  
  for ele, sub_ele, tot_txt, rev_pl in zip(rating_reviews, text_reviews, total_review, review_date) : # Iterating through the list
      Rating.append(ele.text) #Appending to empty list
      Review.append(sub_ele.text)
      Comp_rev.append(tot_txt.text)
      Rev_place.append(rev_pl.text)

#Dataframe Creation by combining collected data and making the column names of our own!
IMDB_rating = pd.DataFrame(list(zip(Rating, Review,Comp_rev, Rev_place )), columns =['Rating', 'Short_Review', 'Total_Review' ,'Posted_Date']) 
print("Total Data Frame Length is {0}".format(len(IMDB_rating)))

IMDB_rating
```

```python
IMDB_rating.to_csv('IMDB_Movie_rating.csv')
```

```python
nltk.download('omw-1.4')
stopword_list = stopwords.words('english')
snow_stemmer = SnowballStemmer("english")
lemmatizer = WordNetLemmatizer()
def clean(IMDB_rating):
    IMDB_rating.loc[:,["clean_txt"]]=IMDB_rating["Short_Review"].apply(lambda ele: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", ele))
    IMDB_rating.loc[:,["clean_txt"]]=IMDB_rating["clean_txt"].apply(lambda ele: re.sub(r"\d+", "", ele))
    IMDB_rating.loc[:,["clean_txt"]]=IMDB_rating["clean_txt"].apply(lambda string: ' '.join([w for w in string.split() if w not in (stopword_list)]))
    IMDB_rating.loc[:,["clean_txt"]]=IMDB_rating["clean_txt"].str.lower()
    IMDB_rating.loc[:,["tokens"]] = IMDB_rating["clean_txt"].apply(lambda x: word_tokenize(x))
    def word_stemmer(text):
      stem_text = [snow_stemmer.stem(i) for i in text]
      return stem_text
    IMDB_rating.loc[:,['text_stem']] = IMDB_rating["tokens"].apply(lambda x: word_stemmer(x))
    def word_lemmatizer(text):
      lem_text = [lemmatizer.lemmatize(i) for i in text]
      return lem_text
    IMDB_rating.loc[:,['text_lem']] = IMDB_rating["tokens"].apply(lambda x: word_lemmatizer(x))
    return IMDB_rating
cleaned_data = clean(IMDB_rating)
cleaned_data["Rating"]=cleaned_data["Rating"].str.replace("\n","")
cleaned_data
```
