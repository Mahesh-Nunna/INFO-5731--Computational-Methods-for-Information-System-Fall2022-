```python
import pandas as pd

data_file="http://www.textfiles.com/food/antibiot.txt"
data=pd.read_csv(data_file,error_bad_lines=False,names=['lines'])
data.head(10)
```

```python
# Number of sentences

data['Number of Sentences'] = data['lines'].apply(lambda x: len(str(x).split(".")))

data[['lines','Number of Sentences']]
```

```python
# Number of Words

data['No of Words']=data['lines'].apply(lambda x: len(x.split(" ")))

data[['lines','No of Words']]
```

```python
# Number of Characters

data['Number of Characters'] = data['lines'].str.len()
data[['lines','Number of Characters']]
```

```python
# Number of Stopwords
from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
stop = stopwords.words('english')

data['Number of Stopwords'] = data['lines'].apply(lambda x: len([x for x in x.split() if x in stop]))
data[['lines','Number of Stopwords']]
```

```python
# Number of Special Characters
data['special character count']=data['lines'].apply(lambda x: len([y for y in x.split() if y.startswith(('#','@'))]))
data[['lines','special character count']]
```

```python
# Number of Numerics
data['numeric count']=data['lines'].apply(lambda x: len([y for y in x.split() if y.isdigit()]))
data[['lines','numeric count']]
```

```python
# Number of uppercase words

data['uppercase words']=data['lines'].apply(lambda x: len([y for y in x.split() if y.isupper()]))
data[['lines','uppercase words']]
```

```python
# Lower casing

data['lines']=data['lines'].str.lower()
data['lines']
```

```python
# Punctuation removal

import string
data['lines']=data['lines'].str.replace('[{}]'.format(string.punctuation), '')
data['lines']
```

```python
# Stopwords removal

from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop=stopwords.words('english')

data['lines']=data['lines'].apply(lambda x:" ".join(x for x in x.split() if x not in stop))
data['lines']
```

```python
# Frequent words removal

freq=pd.Series(' '.join(data['lines']).split()).value_counts()[:10]
freq=list(freq.index)

data['lines']=data['lines'].apply(lambda x:" ".join(x for x in x.split() if x not in freq))
data['lines']
```

```python
# Rare words removal

rare=pd.Series(' '.join(data['lines']).split()).value_counts()[-10:]

rare=list(rare.index)
data['lines']=data['lines'].apply(lambda x:" ".join(x for x in x.split() if x not in rare))
data['lines']
```

```python
# Spelling correction

from textblob import TextBlob

data['lines']=data['lines'].apply(lambda x: str(TextBlob(x).correct()))
data['lines']
```

```python
# Stemming

from nltk.stem import PorterStemmer
stemmer=PorterStemmer()

data['lines']=data['lines'].apply(lambda x: " ".join([stemmer.stem(word) for word in x.split()]))
data['lines']
```
